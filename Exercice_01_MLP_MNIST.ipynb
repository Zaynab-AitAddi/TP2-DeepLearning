## Implémentation d'un Perceptron Multicouche (MLP) sur MNIST

**Objectif :** Implémenter un MLP pour la classification des chiffres manuscrits sur MNIST selon l’architecture imposée.

### Spécifications rappelées
- **Données :** MNIST — images 28×28 (vecteur 784), 60 000 train / 10 000 test, classes 0–9.
- **Architecture :** Entrée 784 → Dense 128 (ReLU) → Dense 64 (ReLU) → Sortie 10 (Softmax).
- **Compilation :** optimiseur Adam, perte `categorical_crossentropy`, métrique précision (`accuracy`).
- **Entraînement :** 5 époques, batch 64, **20 %** du jeu d’entraînement réservés à la validation (`validation_split=0.2`).

**Note :** La perte catégorielle exige des étiquettes au format **one-hot** (`to_categorical`).

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.utils import to_categorical

# Chargement MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Aplatissement 28x28 -> 784 et normalisation [0, 1]
x_train = x_train.reshape(-1, 784).astype("float32") / 255.0
x_test = x_test.reshape(-1, 784).astype("float32") / 255.0

# Encodage one-hot pour categorical_crossentropy
y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)

# Architecture imposée
model = Sequential(
    [
        Input(shape=(784,)),
        Dense(128, activation="relu"),
        Dense(64, activation="relu"),
        Dense(10, activation="softmax"),
    ]
)

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"],
)

model.summary()

history = model.fit(
    x_train,
    y_train_cat,
    epochs=5,
    batch_size=64,
    validation_split=0.2,
    verbose=1,
)

test_loss, test_acc = model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Perte (test) : {test_loss:.4f} — Précision (test) : {test_acc:.4f}")
```
