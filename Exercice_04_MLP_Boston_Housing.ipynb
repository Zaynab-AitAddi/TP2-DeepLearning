## MLP pour la régression — Boston Housing

**Objectif :** Prédire le **prix médian des logements** (en milliers de dollars) à partir de **13** caractéristiques avec un MLP en **régression**.

**Compilation imposée :** Adam, perte **MSE** (`mean_squared_error`), métrique **RMSE** (erreur quadratique moyenne en racine).

### Spécifications rappelées
- **Données :** Boston Housing — **506** maisons, **13** attributs ; cible continue (prix médian).
- **Architecture :** Entrée 13 → Dense 64 (ReLU) → Dense 32 (ReLU) → Sortie **1** neurone (**activation linéaire**, adaptée à une valeur réelle).
- **Compilation :** Adam, `mean_squared_error`, métrique **`RootMeanSquaredError`** (RMSE).
- **Entraînement :** **50** époques, batch **32**.
- Le module `boston_housing` de Keras fournit déjà des jeux **train** et **test** ; on **normalise** les entrées avec la moyenne et l’écart-type calculés **sur le train uniquement**.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import boston_housing
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

(x_train, y_train), (x_test, y_test) = boston_housing.load_data()

mean = x_train.mean(axis=0)
std = x_train.std(axis=0)
std_safe = np.where(std == 0, 1e-8, std)
x_train = ((x_train - mean) / std_safe).astype("float32")
x_test = ((x_test - mean) / std_safe).astype("float32")

model = Sequential(
    [
        Input(shape=(13,)),
        Dense(64, activation="relu"),
        Dense(32, activation="relu"),
        Dense(1),
    ]
)

model.compile(
    optimizer="adam",
    loss="mean_squared_error",
    metrics=[tf.keras.metrics.RootMeanSquaredError(name="rmse")],
)

model.summary()

model.fit(x_train, y_train, epochs=50, batch_size=32, verbose=1)

test_loss, test_rmse = model.evaluate(x_test, y_test, verbose=0)
print(f"Perte MSE (test) : {test_loss:.4f} — RMSE (test) : {test_rmse:.4f}")
```
